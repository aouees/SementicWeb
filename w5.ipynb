```python
from rdflib import Namespace,Literal,Graph,URIRef,BNode
from rdflib.namespace import XSD,DCAT,DCTERMS,FOAF,RDF
```

```python
EX=Namespace('https//:example.org')
```

```python
g=Graph()
robertMueller=BNode()
```

```python
g.bind('ex',EX)
g.bind('foaf',FOAF)
g.bind('rdf',RDF)
position=Literal('Director of the federal ...')
```

```python
g.add((robertMueller,EX['position_head'],position))
g.add((robertMueller,FOAF['name'],Literal('robert muller')))
g.add((robertMueller,RDF.type,EX['human']))

```

```python
print(g.serialize(format='ttl'))
```

```python
import requests
from IPython.display import Image
params = {
    'rdf': g.serialize(format='ttl'),
    'from': 'ttl',
    'to':'png'
    }
response = requests.get('http://www.ldf.fi/service/rdf-grapher',params=params)
img = Image(response.content)
display(img)
```
